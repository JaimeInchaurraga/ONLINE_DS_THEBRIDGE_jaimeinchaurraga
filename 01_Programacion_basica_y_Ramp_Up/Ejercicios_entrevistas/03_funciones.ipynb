### Ejercicio: Función para Calcular el Factorial de un Número

**Enunciado:**

Vas a crear una función en Python que calcule el factorial de un número entero no negativo.  
 El factorial de un número `n` se define como el producto de todos los enteros positivos desde 1 hasta `n`. Se denota como `n!`. Por ejemplo, el factorial de 5 es `5! = 5 × 4 × 3 × 2 × 1 = 120`.

**Instrucciones**:

1. **Definir la función**:
   - Crea una función llamada `calcular_factorial` que tome un único argumento `n`.
   - La función debe verificar que `n` sea un número entero no negativo.
   - Si `n` no es un entero no negativo, la función debe devolver un mensaje indicando que la entrada no es válida.
   
2. **Calcular el factorial**:
   - Si `n` es 0, la función debe devolver 1, ya que `0! = 1`.
   - Para otros valores de `n`, calcula el factorial usando un bucle `for` o `while`.

3. **Probar la función**:
   - Llama a la función `calcular_factorial` con varios valores de prueba y muestra los resultados.

4. **Ejemplo de ejecución**:
   - Llamar a `calcular_factorial(5)` debería devolver 120.
   - Llamar a `calcular_factorial(0)` debería devolver 1.
   - Llamar a `calcular_factorial(-3)` debería devolver un mensaje de error.

### Consideraciones:
- Asegúrate de manejar adecuadamente las entradas del usuario.
- Usa un bucle para calcular el factorial.
- No uses funciones predefinidas para calcular el factorial (como `math.factorial`), ya que el objetivo es practicar la lógica de programación.

Este ejercicio te ayudará a practicar la creación de funciones, el uso de bucles, y la validación de entradas en Python.


```python
def calcular_factorial(n):
    # Verificar si la entrada es un entero no negativo
    if not isinstance(n, int) or n < 0:
        return "La entrada no es válida. Por favor, introduce un entero no negativo."
    
    # Caso especial para 0! = 1
    if n == 0:
        return 1
    
    # Calcular el factorial usando un bucle
    factorial = 1
    for i in range(1, n + 1):
        factorial *= i
    
    return factorial

# Probar la función con varios valores
print(f"Factorial de 5: {calcular_factorial(5)}")  # Debería devolver 120
print(f"Factorial de 0: {calcular_factorial(0)}")  # Debería devolver 1
print(f"Factorial de -3: {calcular_factorial(-3)}")  # Debería devolver un mensaje de error
print(f"Factorial de 7: {calcular_factorial(7)}")  # Debería devolver 5040
print(f"Factorial de 1: {calcular_factorial(0)}")  # Debería devolver 1
```

### Ejercicio: Crear una Agenda de Contactos

**Enunciado:**

Vas a crear una agenda de contactos utilizando un diccionario en Python. La agenda permitirá agregar, eliminar, buscar y mostrar contactos. Cada contacto tendrá un nombre y un número de teléfono. Implementa las siguientes funcionalidades:

**Instrucciones**:

1. **Definir las funciones**:
   - `agregar_contacto(agenda, nombre, telefono)`: Esta función debe agregar un contacto a la agenda.
   - `eliminar_contacto(agenda, nombre)`: Esta función debe eliminar un contacto de la agenda.
   - `buscar_contacto(agenda, nombre)`: Esta función debe buscar un contacto por nombre y devolver su número de teléfono.
   - `mostrar_agenda(agenda)`: Esta función debe mostrar todos los contactos en la agenda.

2. **Implementar el menú**:
   - Muestra un menú con las siguientes opciones:
     1. Agregar contacto.
     2. Eliminar contacto.
     3. Buscar contacto.
     4. Mostrar agenda.
     5. Salir.

3. **Gestión del menú**:
   - Utiliza un bucle `while` para mostrar el menú y solicitar la opción del usuario hasta que elija salir.
   - Llama a la función correspondiente según la opción seleccionada por el usuario.

4. **Validación de entradas**:
   - Asegúrate de validar las entradas del usuario para evitar errores.

### Consideraciones:
- Usa un diccionario para almacenar los contactos, donde las claves son los nombres y los valores son los números de teléfono.
- Maneja adecuadamente las entradas del usuario para evitar errores.
- Proporciona mensajes adecuados al usuario para confirmar las acciones realizadas (por ejemplo, contacto agregado, contacto eliminado, etc.).

### Ejemplo de ejecución:

- Agregar un contacto: el usuario introduce el nombre y el número de teléfono.
- Eliminar un contacto: el usuario introduce el nombre del contacto a eliminar.
- Buscar un contacto: el usuario introduce el nombre del contacto a buscar.
- Mostrar la agenda: se muestran todos los contactos en la agenda.
- Salir: se termina el programa con un mensaje de despedida.

Este ejercicio te ayudará a practicar el uso de diccionarios, la creación de funciones, la gestión de menús y la validación de entradas en Python.


```python
def agregar_contacto(agenda, nombre, telefono):
    agenda[nombre] = telefono
    print(f"Contacto {nombre} agregado con éxito.")

def eliminar_contacto(agenda, nombre):
    if nombre in agenda:
        del agenda[nombre]
        print(f"Contacto {nombre} eliminado con éxito.")
    else:
        print(f"Contacto {nombre} no encontrado en la agenda.")

def buscar_contacto(agenda, nombre):
    if nombre in agenda:
        return f"El número de {nombre} es {agenda[nombre]}"
    else:
        return f"Contacto {nombre} no encontrado en la agenda."

def mostrar_agenda(agenda):
    if agenda:
        print("Agenda de contactos:")
        for nombre, telefono in agenda.items():
            print(f"{nombre}: {telefono}")
    else:
        print("La agenda está vacía.")

def main():
    agenda = {}
    opcion = 0

    print("Bienvenido a la agenda de contactos.")
    while opcion != 5:
        print("\nPor favor, selecciona una opción:")
        print("1. Agregar contacto")
        print("2. Eliminar contacto")
        print("3. Buscar contacto")
        print("4. Mostrar agenda")
        print("5. Salir")
        
        try:
            opcion = int(input("Escribe el número de la opción: "))
        except ValueError:
            print("Por favor, introduce un número válido.")
            continue

        if opcion == 1:
            nombre = input("Introduce el nombre del contacto: ")
            telefono = input("Introduce el número de teléfono del contacto: ")
            agregar_contacto(agenda, nombre, telefono)
        
        elif opcion == 2:
            nombre = input("Introduce el nombre del contacto a eliminar: ")
            eliminar_contacto(agenda, nombre)
        
        elif opcion == 3:
            nombre = input("Introduce el nombre del contacto a buscar: ")
            print(buscar_contacto(agenda, nombre))
        
        elif opcion == 4:
            mostrar_agenda(agenda)
        
        elif opcion == 5:
            print("Gracias por usar la agenda de contactos. ¡Hasta luego!")
        else:
            print("Por favor, selecciona una opción válida (1-5).")

if __name__ == "__main__":
    main()

```
