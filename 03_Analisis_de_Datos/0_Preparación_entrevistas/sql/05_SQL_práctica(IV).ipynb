```python
import pandas as pd
import sqlite3
```

```python
# Conectamos con la base de datos chinook.db

connection = sqlite3.connect("./data_tres/northwind.db")

# Obtenemos un cursor que utilizaremos para hacer las queries

cursor_clase = connection.cursor()
```

```python
# Con esta función leemos los datos y lo pasamos a un DataFrame de Pandas
def sql_query(query):

    # Ejecuta la query
    cursor_clase.execute(query)
                                    

    # Almacena los datos de la query 
    ans = cursor_clase.fetchall()

    # Obtenemos los nombres de las columnas de la tabla
    names = [description[0] for description in cursor_clase.description]

    return pd.DataFrame(ans,columns=names)
```

```python
# ¿Cuáles son los clientes más valiosos en términos de ventas totales?
query = '''
SELECT Customers.ContactName AS Customer, SUM("Order Details".UnitPrice) AS Total
FROM customers
JOIN orders 
ON Customers.CustomerID = Orders.CustomerID
JOIN "Order Details" 
ON Orders.OrderID = "Order Details".OrderID
GROUP BY Customers.ContactName
ORDER BY Total DESC
LIMIT 5
'''

sql_query(query)
```

```python
# ¿Cuál es el promedio de ventas por cliente?
query = '''
SELECT Customers.CustomerID, Customers.ContactName, AVG("Order Details".UnitPrice * "Order Details".Quantity) AS AvgSales
FROM Customers
JOIN Orders 
ON Customers.CustomerID = Orders.CustomerID
JOIN "Order Details" 
ON Orders.OrderID = "Order Details".OrderID
GROUP BY Customers.CustomerID, Customers.ContactName
ORDER BY AvgSales DESC

'''

sql_query(query)
```

```python
# ¿Cuál es el total de ventas por cliente?
query = '''
SELECT Customers.CustomerID, Customers.ContactName, SUM("Order Details".UnitPrice * "Order Details".Quantity) AS TotalSales
FROM Customers
JOIN Orders 
ON Customers.CustomerID = Orders.CustomerID
JOIN "Order Details" 
ON Orders.OrderID = "Order Details".OrderID
GROUP BY Customers.CustomerID, Customers.ContactName
ORDER BY TotalSales DESC

'''

sql_query(query)
```

```python
# ¿Cuáles son los productos más vendidos?

query = '''
SELECT products.ProductName, SUM("order details".Quantity) AS TotalQuantity
FROM products
JOIN "order details"
ON products.ProductID = "order details".ProductID
GROUP BY products.ProductName
ORDER BY TotalQuantity DESC
LIMIT 10

'''

sql_query(query)
```
