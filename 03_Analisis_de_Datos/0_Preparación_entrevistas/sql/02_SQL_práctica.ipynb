```python
import pandas as pd
import sqlite3 as sq
```

```python
# Conectamos con la base de datos chinook.db

connection = sq.connect("./data_dos/chinook_dos.db")

# Obtenemos un cursor que utilizaremos para hacer las queries

cursor_clase = connection.cursor()
```

```python
def sql_query(query):

    # Ejecuta la query
    cursor_clase.execute(query)
                                    

    # Almacena los datos de la query 
    ans = cursor_clase.fetchall()

    # Obtenemos los nombres de las columnas de la tabla
    names = [description[0] for description in cursor_clase.description]

    return pd.DataFrame(ans,columns=names)
```

```python

def get_column_names(table_name):
    # Conectar a la base de datos
    connection = sq.connect("./data_dos/chinook_dos.db")
    cursor = connection.cursor()

    # Ejecutar PRAGMA table_info para obtener la información de las columnas
    cursor.execute(f'PRAGMA table_info({table_name});')
    columns_info = cursor.fetchall()

    # Cerrar la conexión
    connection.close()

    # Extraer y devolver los nombres de las columnas
    column_names = [col[1] for col in columns_info]
    return column_names

# Obtener los nombres de las columnas de la tabla 'customers'
columns = get_column_names('customers')
print("Column names in 'customers' table:", columns)
```

```python
# 5. Selecciona los clientes cuyo apellido sea 'Tremblay'. --> cambio de apellido porque del otro no hay clientes

query = '''
SELECT *
FROM customers
WHERE LastName = 'Tremblay'
'''
sql_query(query)
```

```python
# 6. Selecciona los clientes que viven en el Reino Unido (UK) y ordena los resultados por `FirstName`.

query = '''
SELECT *
FROM customers
WHERE Country = 'United Kingdom'
ORDER BY FirstName
'''
sql_query(query)
```

```python
# EJERCICIO DOS
# APARTADO TRES
# 7. Encuentra el número total de clientes.

query = '''
SELECT COUNT (*) AS TotalCustomers
FROM customers
'''
sql_query(query)
```

```python
# 8. Calcula la cantidad total de ventas (`Total`) en la tabla `invoices`.

query = '''
SELECT SUM(Total) AS TotalSales
FROM invoices
'''
sql_query(query)
```

```python
# 9. Encuentra el número de clientes por país en la tabla `customers`.

query = '''
SELECT Country, COUNT (*)
FROM Customers
GROUP BY Country
'''
sql_query(query)
```

```python
# APARTADO 4
# 10. Selecciona todos los registros de la tabla `invoices` junto con los nombres de los clientes correspondientes (`FirstName`, `LastName`) de la tabla `customers`.

query = '''
SELECT invoices.*, customers.FirstName, customers.LastName
FROM invoices
JOIN customers ON invoices.CustomerId = customers.CustomerId
LIMIT 5;
'''
sql_query(query)

```

```python
# 11. Selecciona los nombres de los artistas (`Name`) y los nombres de los álbumes (`Title`) de la tabla `albums`.
```

```python

query = '''
SELECT artists.Name AS ArtistName, albums.Title AS AlbumTitle
FROM albums
JOIN artists ON albums.ArtistId = artists.ArtistId
ORDER BY ArtistName ASC
LIMIT 5
'''
sql_query(query)
```

```python
# 12. Selecciona los nombres de las pistas (`Name`) junto con los nombres de los géneros (`Name`) correspondientes de la tabla `tracks` y `genres`.
```

```python

query = '''
SELECT genres.Name AS Genre, tracks.Name AS Track
FROM genres
RIGHT JOIN tracks ON genres.GenreId = tracks.GenreId
LIMIT 5
'''
sql_query(query)
```
